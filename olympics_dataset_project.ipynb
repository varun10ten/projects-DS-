```python
def extract_medal(year,country):
    flag=0
    if year=='overall' and country == 'overall':
        temp_df=medal_df
    if year=='overall' and country!='overall':
        flag=1
        temp_df=medal_df[medal_df['region']==country]
    if year!='overall' and country == 'overall':
        temp_df=medal_df[medal_df['Year']==int(year)]
    if year!='overall' and country!='overall':
        temp_df=medal_df[(medal_df['Year']==year)&(medal_df['region']==country)]
    if flag==1:
        x=temp_df.groupby('Year').sum()[['Gold','Silver','Bronze']].sort_values('Gold',ascending=False).reset_index()
    else:
        x=temp_df.groupby('region').sum()[['Gold','Silver','Bronze']].sort_values('Gold',ascending=False).reset_index()
    x['Total']=x['Gold']+x['Silver']+x['Bronze']
    print(x)
```

finding correlation matrix of the dataset as to find the relation between datsets.

```python
x=df.drop_duplicates(['Year','Sport','Event'])
```

```python
# Import necessary libraries and set inline plotting
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'x' is your DataFrame or pivot table
pivot_table = x.pivot_table(index='Sport', columns='Year', values='Event', aggfunc='count').fillna(0).astype(int)

# Create a heatmap with annotations
plt.figure(figsize=(25, 25))
sns.heatmap(pivot_table, annot=True, fmt='d', cmap='YlGnBu')  # Use 'fmt='d'' for integer formatting
plt.title('Event Count by Sport and Year')
plt.xlabel('Year')
plt.ylabel('Sport')
plt.show()
```

```python
temp_df=df.dropna(subset='Medal')
temp_df
```

```python
def most_medals_athletes(df, sport):
    temp_df = df.dropna(subset=['Medal'])
    if sport != 'overall':
        temp_df = temp_df[temp_df['Sport'] == sport]
    name_counts = temp_df['Name'].value_counts().reset_index()
    #want to rename count column into medal count column
    name_counts.columns = ['Name', 'Medal Count']
# Merge with original DataFrame to add 'region' column
    result_df = name_counts.merge(df[['Name', 'region','Sport']], on='Name', how='left').drop_duplicates('Name')

    return result_df
```

```python
most_medals_athletes(df, 'overall')  
```
