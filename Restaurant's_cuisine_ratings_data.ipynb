```python
import seaborn as sns

# List of columns for which you want to create box plots
columns_to_plot = ['Food Rating', 'Service Rating', 'Overall Rating']

# Create a single figure with multiple box plots using Seaborn
plt.figure(figsize=(10, 6))
ax = sns.boxplot(data=df[columns_to_plot], orient='h', notch=True, palette='Set3')

# Customize the plot
plt.title('Rating Box Plots', fontsize=16)
plt.xlabel('Values', fontsize=14)
plt.yticks(fontsize=12)

# Add annotations to the box plots
medians = df[columns_to_plot].median().tolist()
q1 = df[columns_to_plot].quantile(0.25).tolist()
q3 = df[columns_to_plot].quantile(0.75).tolist()

for i, box in enumerate(ax.artists):
    # Get the current box color from the palette
    box_color = sns.color_palette('Set3')[i % len(columns_to_plot)]

    box.set_facecolor(box_color)  # Apply box colors

    # Add median annotation
    ax.text(
        medians[i],
        i + 0.1,
        f'Median: {medians[i]:.2f}',
        fontsize=10,
        color='red',
        ha='center',
        va='center',
        fontweight='bold',
    )

    # Add Q1 and Q3 annotations
    ax.text(
        q1[i],
        i - 0.1,
        f'Q1: {q1[i]:.2f}',
        fontsize=10,
        color='blue',
        ha='center',
        va='center',
        fontweight='bold',
    )
    ax.text(
        q3[i],
        i + 0.3,
        f'Q3: {q3[i]:.2f}',
        fontsize=10,
        color='blue',
        ha='center',
        va='center',
        fontweight='bold',
    )

# Save the figure as an image file (e.g., PNG)
plt.savefig('rating_boxplots.png', bbox_inches='tight')

# Show the plot
plt.show()
```

## Analysis of Qulatitave Data

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def plot_categorical_ratio(df, categorical_column, save_path=None):
    # Calculate the ratio of each category
    category_ratios = df[categorical_column].value_counts(normalize=True).reset_index()
    category_ratios.columns = [categorical_column, 'Ratio']

    # Sort the DataFrame by ratio (optional)
    category_ratios = category_ratios.sort_values(by='Ratio', ascending=False)

    # Calculate the total count of observations
    total_count = df[categorical_column].count()

    # Set a custom color palette
    custom_palette = sns.color_palette('pastel')

    # Create a bar chart using Seaborn with customizations
    plt.figure(figsize=(10, 6))
    ax = sns.barplot(data=category_ratios, x=categorical_column, y='Ratio', palette=custom_palette)
    plt.xticks(rotation=45)  # Rotate x-axis labels if needed

    # Display individual counts above each bar
    for p in ax.patches:
        ax.annotate(f'{p.get_height()*total_count:.0f}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='bottom', fontsize=12, color='black')

    # Customize the plot
    plt.title(f'Ratios of {categorical_column} (Total Count: {total_count})', fontsize=16)
    plt.xlabel(categorical_column, fontsize=14)
    plt.ylabel('Ratio', fontsize=14)
    plt.grid(axis='y', linestyle='--', alpha=0.7)

    # Save the plot as an image (optional)
    if save_path:
        plt.savefig(save_path, bbox_inches='tight')

    # Show the plot
    plt.show()

# Example usage:
# Load your dataset (replace 'your_dataset.csv' with the actual file path)
# df = pd.read_csv('your_dataset.csv')

# Plot the categorical ratio for a specific column and save it as an image
# plot_categorical_ratio(df, 'your_categorical_column', save_path='output.png')  # Replace with the actual column name and file path
```
